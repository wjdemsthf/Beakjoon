#Shortest Path Algorithm Comparison
- Comparison of runtime

```python
#@title Complete Code
# Shortest Path Algorithm Comparison (C언어)

%%writefile Shortest_path.c

#include <stdio.h>
#include <stdlib.h>
#include <time.h>

#define n_MAX 250
#define w_MAX 10
#define infinite w_MAX + 100

void floyd(int n, int W[][n_MAX], int D[][n_MAX], int P[][n_MAX]);
void dijkstra(int n, int W[][n_MAX], int D[], int P[]);
void path(int P[][n_MAX], int start, int end);

int main() {
   int vertex, density, start, end;
   int W[n_MAX][n_MAX], D[n_MAX][n_MAX], P[n_MAX][n_MAX];
   float res[2] = { 0.0 };
   clock_t clock_S, clock_E;

   while (1) {
      printf("Enter Number of Vertex(1 ~ %d): ", n_MAX);
      scanf("%d", &vertex);
      if (vertex > 0 && vertex <= n_MAX) break;
      else printf("Error: Out of Range\n");
   }
   while (1) {
      printf("Enter Ratio of Density(0 ~ 100): ");
      scanf("%d", &density);
      if (density >= 0 && density <= 100) break;
      else printf("Error: Out of Range\n");
   }
   srand(time(NULL));
   for (int i = 0; i < vertex; i++) {
      W[i][i] = 0;
      for (int j = i + 1; j < vertex; j++) {
         if (rand() % 100 + 1 <= density)
            W[i][j] = W[j][i] = rand() % w_MAX + 1;
         else
            W[i][j] = W[j][i] = infinite;
      }
   }
   printf("\n\n");
   printf("======================== W ========================\n");
   for (int i = 0; i < vertex; i++) {
      for (int j = 0; j < vertex; j++)
         printf("%3d  ", (W[i][j] != infinite) ? W[i][j] : -1);
      printf("\n");
   }
   printf("===================================================\n\n");
   for (int i = 0; i < vertex; i++)
      for (int j = 0; j < vertex; j++)
         P[i][j] = -1;

   printf("====================== Floyd ======================\n\n");
   while (1) {
      printf("Enter Start Vertex Index(0 ~ %d): ", vertex - 1);
      scanf("%d", &start);
      if (start >= 0 && start < vertex) break;
      else printf("Error: Out of Range\n");
   }
   while (1) {
      printf("Enter End Vertex Index(0 ~ %d): ", vertex - 1);
      scanf("%d", &end);
      if (end >= 0 && end < vertex) break;
      else printf("Error: Out of Range\n");
   }

   clock_S = clock();
   floyd(vertex, W, D, P);
   clock_E = clock();
   res[0] = (float)(clock_E - clock_S) / CLOCKS_PER_SEC;
   
   printf("\n");
   if (D[start][end] != infinite) {
      printf("Length of Vertex_%d to Vertex_%d: %d\n", start, end, D[start][end]);
      printf("Path: V%d -> ", start);
      path(P, start, end);
      if (start != end)
         printf("V%d", end);
   }
   else {
      printf("Length of Vertex_%d to Vertex_%d: Infinite\n", start, end);
      printf("Path: - ");
   }
   printf("\n\n===================================================\n\n");
   
   printf("===================== Dijkstra ====================\n\n");
   printf("Start vertex Index of Dijkstra Function fixed to 0\n");
   while (1) {
      printf("Enter End Vertex Index(0 ~ %d): ", vertex - 1);
      scanf("%d", &end);
      if (end >= 0 && end < vertex) break;
      else printf("Error: Out of Range\n");
   }

   clock_S = clock();
   dijkstra(vertex, W, D[0], P[0]);
   clock_E = clock();
   res[1] = (float)(clock_E - clock_S) / CLOCKS_PER_SEC;

   printf("\n");
   if (D[0][end] != infinite) {
      printf("Length of Vertex_0 to Vertex_%d: %d\n", end, D[0][end]);
      printf("Path: V0 -> ");
      path(P, 0, end);
      if (0 != end)
         printf("V%d", end);
   }
   else {
      printf("Length of Vertex_0 to Vertex_%d: Infinite\n", end);
      printf("Path: - ");
   }
   printf("\n\n===================================================\n\n");

   printf("Floyd    Algorithm: %fsec\n", res[0]);
   printf("Dijkstra Algorithm: %fsec\n", res[1]);

   return 0;
}

void floyd(int n, int W[][n_MAX], int D[][n_MAX], int P[][n_MAX]) {
   for (int i = 0; i < n; i++)
      for (int j = 0; j < n; j++)
         D[i][j] = W[i][j];

   for (int k = 0; k < n; k++)
      for (int i = 0; i < n; i++)
         for (int j = 0; j < n; j++)
            if (D[i][j] > D[i][k] + D[k][j]) {
               P[i][j] = k;
               D[i][j] = D[i][k] + D[k][j];
            }
}

void dijkstra(int n, int W[][n_MAX], int D[], int P[]) {
   int length[n_MAX], tmp;

   D[0] = 0;
   for (int i = 0; i < n; i++)
      length[i] = W[0][i];

   for (int i = 1; i < n; i++) {
      tmp = 0;
      for (int j = 1; j < n; j++) {
         if (tmp == 0 && length[j] < infinite && length[j] != 0)
            tmp = j;
         else if (length[tmp] > length[j] && length[j] != 0)
            tmp = j;
      }
      D[tmp] = length[tmp];
      for (int j = 1; j < n; j++) {
         if (length[j] > length[tmp] + W[tmp][j] && length[j] != 0) {
            P[j] = tmp;
            length[j] = length[tmp] + W[tmp][j];
         }
      }
      length[tmp] = 0;
   }
}

void path(int P[][n_MAX], int start, int end) {
   if (P[start][end] != -1) {
      path(P, start, P[start][end]);
      printf("V%d -> ", P[start][end]);
      path(P, P[start][end], end);
   }
}
```
